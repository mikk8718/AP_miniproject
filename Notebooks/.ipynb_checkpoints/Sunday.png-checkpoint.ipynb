```python
from scipy import *
from numpy import *
from thinkdsp import *
import matplotlib.pyplot as plt
import math
import IPython.display as ipd
```

```python
sampleFreq = 44100
time = arange(0, 44100 * 2).T / sampleFreq
y = sin(2*pi*200*time) + sin(2*pi*500*time)
```

```python
n = len(y)
d = 1 / 44100
```

```python
hs = fft.fft(y)
fs = fft.fftfreq(n, d)
```

```python
amps = abs(hs)
```

```python
plt.plot(fs, amps / max(amps))
plt.xlim(150, 550)
plt.grid()
```

```python
ipd.Audio(y, rate=44100)
```

```python
print(abs(hs[2]))
```

```python
def lowpass(cutoff):
    hs[abs(fs) > cutoff] = 0
```

```python
lowpass(200)
```

```python
amps = abs(hs)
```

```python
plt.plot(fs, amps / max(amps))
plt.grid()
plt.xlim(150, 550)
```
