```python
import thinkdsp
from IPython.display import display
from ipywidgets import interact, fixed
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
from scipy.fftpack import rfft, irfft, fftfreq, fft, rfftfreq, ifft
import sounddevice as sd
import scipy
```

```python
signal = thinkdsp.SinSignal(freq=500, amp=1)
signal_2 = thinkdsp.SinSignal(freq=200, amp=1)
signal_3 = signal + signal_2
wave = signal_3.make_wave(duration=5, start=0, framerate=44100)
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.low_pass(400)
spectrum.plot()
plt.xlim(0, 1000)
```

```python
sd.play(wave.ys.real, 44100)
sd.wait()
```
